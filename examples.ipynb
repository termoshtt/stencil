```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

# Diffusion equation

```python
! cargo run --release --example heat
```

```python
df = pd.read_json("heat_periodic.json", lines=True)
```

```python
dt = 1e-3
step = 1
T = len(df) // 5
for t in range(5):
    plt.plot(df.iloc[t*T], label="t={:.01f}".format(T*t * dt * step))
plt.legend()
```
