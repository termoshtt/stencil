# KdV equation

```python
! cargo run --release --example heat
```

```python
df = pd.read_json("kdv.json", lines=True)
```

```python
df[0].plot()
```

```python
dt = 1e-5
step = 1000
T = len(df) // 4
for t in range(4):
    plt.plot(df.iloc[t*T], label="t={:.01f}".format(T*t * dt * step))
plt.legend()
```
